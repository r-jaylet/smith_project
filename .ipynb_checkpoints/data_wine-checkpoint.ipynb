```python
import requests
import json
import pandas as pd
import re
```

```python
wine_category = ['reds', 'whites', 'sparkling', 'rose']

def wine_cat_info(cat):

    url = 'https://api.sampleapis.com/wines/'
    
    selected_category = cat
    response = requests.get(url + selected_category)
    response_json = json.loads(response.text)
    
    return response_json
```

```python
def format_selected_wine(selected_wine, cat):
    
    selection = {}

    if any(char.isdigit() for char in selected_wine['wine']):
        selection['name'] = selected_wine['wine'].replace(' ' + re.findall(r'\d+', selected_wine['wine'])[0], '')
        selection['vintage'] = int(re.findall(r'\d+', selected_wine['wine'])[0])
    else:
        selection['name'] = selected_wine['wine']
        selection['vintage'] = 0

    selection['winery'] = selected_wine['winery']
    selection['rating'] = selected_wine['rating']['average']
    selection['country'] = selected_wine['location'].split('\n·\n')[0]
    if len(selected_wine['location'].split('\n·\n')) > 1:
            selection['city'] =  selected_wine['location'].split('\n·\n')[1]
    else:
            selection['city'] = None

    selection['image'] = selected_wine['image']
    selection['degree'] = '13%'
    selection['category'] = cat

    return selection
```

```python
wine_list = []

for cat in wine_category:
    
    response_json = wine_cat_info(cat)

    for i in range(len(response_json)):
        selection = format_selected_wine(response_json[i], cat)
        wine_list.append(selection)
```

```python
data = pd.DataFrame.from_dict(wine_list, orient='columns')
```
